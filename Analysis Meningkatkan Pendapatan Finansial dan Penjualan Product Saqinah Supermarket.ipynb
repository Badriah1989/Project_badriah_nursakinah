## Latar Belakang Masalah

"Analisis Meningkatkan Pendapatan Finansial Dari Penjualan Produk di Saqinah Supermarket"

Dalam industri Retail, Saqinah Market adalah salah satu retailer terbesar dan populer di kotanya yang memilki 3 cabang di kota yang berbeda yaitu Mandalay, Naypytay,dan Yangon. Sebagai penyedia layanan jasa dalam penjualan produk yang beragam, Saqinah Market menawarkan berbagai produk kepada pelanggan di seluruh kota-kota besar. Dalam konteks ini, penting untuk melakukan analisis keberhasilan keuangan dari penjualan produk guna memahami kinerja bisnis dan mengoptimalkan strategi penjualan.

## Problem Statement (Analisis SMART)

Dari analisa Prblem Statement ini adapun beberapa poin yang akan di berika detail penjelsan diantaranya adalah

Adapun tujuan spesifik:
Meningkatkan pendapatan Saqinah Market sebanyak 40% dengan menganalisis kota yang memiliki quantity penjualan paling besar.

Kriteria SMART:

- Measureable: Peningkatan pendapatan Saqinah Market sebanyak 40%.
- Achievable: Tujuan tersebut dapat dicapai karena Saqinah Market sudah memiliki tim yang dapat meneliti perilaku pelanggan 
- Relevant: Peningkatan pendapatan dan peningkatan penjualan produk merupakan tujuan yang relevan dengan goal dan pencapaian untuk menjadi perusahaan retail terbaik dalam jangka waktu yang lama.
- Time-Bound: Tujuan tersebut akan dicapai dalam waktu satu tahun sejak tahun 2019.

```python
#Membuat visualisasi untuk memberikan informasi gross income

import pandas as pd
import matplotlib.pyplot as plt

# Membaca file data
data = pd.read_csv('supermarket_sales.csv')

# Mengubah kolom "Date" menjadi tipe data datetime
data['Date'] = pd.to_datetime(data['Date'])

# Menghitung keuntungan per produk
data['Profit'] = data['Unit price'] * data['Quantity'] - data['gross income']
profit_per_product = data.groupby('Product line')['Profit'].sum()

# Menentukan produk dengan keuntungan terbesar
most_profitable_product = profit_per_product.idxmax()

# Mengambil data produk yang memberikan keuntungan terbesar
most_profitable_data = data[data['Product line'] == most_profitable_product]

# Mengelompokkan data per bulan
most_profitable_data['Bulan'] = most_profitable_data['Date'].dt.to_period('M')
profit_per_month = most_profitable_data.groupby('Bulan')['Profit'].sum()

# Membuat diagram line chart
profit_per_month.plot(kind='line', figsize=(10, 6))
plt.xlabel('Bulan')
plt.ylabel('Keuntungan')
plt.title('Keuntungan Produk Terlaris per Bulan')
plt.xticks(rotation=45)
plt.show()

print('Produk dengan keuntungan terbesar:', most_profitable_product)
```

## Analisa Skewness
Dalam pengujian skewness sebelum melakukan pengujian statisti baik itu pengujian statistik descriptive maupun pengujian statistik inferensial itu sangat penting dilakukan. Tujuan dari pengujian skewness ini untuk dapat mengetahui apakah data yang akan kita lakukan pengujian dengan uji statistik bersifat data yang berdistribusi normal atau data yang berdistribusi tidak normal, setelah kita mengetahui sifat data tersebut barulah kita dapat menentukan penggunaan methode yang tepat dari analisa uji statistk tersebut.Adapun salah satu syarat untuk dapat dilakukan pengujian skewness dari data yang kita miliki pilihlah beberapa variabel yang bersifat numerik 

```python
##Untuk mengetahu nilai skewness antara variabel quantity,gross income dan total

import pandas as pd

# Membaca file data
data = pd.read_csv('supermarket_sales.csv')

# Membersihkan data dari missing values
clean = data.dropna()

# Mengubah tipe data kolom 'Quantity', 'gross income', dan 'Total' menjadi numerik
clean['Quantity'] = pd.to_numeric(clean['Quantity'], errors='coerce')
clean['gross income'] = pd.to_numeric(clean['gross income'], errors='coerce')
clean['Total'] = pd.to_numeric(clean['Total'], errors='coerce')

# Menghitung kemiringan (skewness) dari variabel 'Quantity', 'gross income', dan 'Total'
skewness = clean[['Quantity', 'gross income', 'Total']].skew()

# Menampilkan hasil kemiringan
print(skewness)
```

Dapat dilihat dari distribusi data pada gross income dan Total memiliki nilai > 0,5 maka dapat dikatakan memiliki sifat highly skew (skew>0,5) dan untuk distribusi data pada Quantity yang memilki nilai skew 0,012 yang dianggap nilai skew nya terlalu kecil maka data ditakatakan masih berdistribusi normal

## Analisa Outlier

Dari hasil analisa skewness pada proses sebelumnya maka dapat disimpulkan bahwa 2 variabel yang memiliki nilai yang bersifat skewness adalah gross income dan Total maka maka perlu dilakukan pengecekan outlier dengan menggunakan methode Turkey Rule karena nilai skewness > 0,5


```python
##Untuk mengetahui nilai outlier pada variabel Gross income

import pandas as pd

# Membaca file data
data = pd.read_csv('supermarket_sales.csv')

# Membersihkan data dari missing values
clean = data.dropna()

# Mengubah tipe data kolom 'Gross income' menjadi numerik
clean['gross income'] = pd.to_numeric(clean['gross income'], errors='coerce')

# Menghitung quartile 1, quartile 3, dan IQR
q1 = clean['gross income'].quantile(0.25)
q3 = clean['gross income'].quantile(0.75)
iqr = q3 - q1

# Menghitung batas atas dan batas bawah
upbound = q3 + 1.5 * iqr
lowbound = q1 - 1.5 * iqr

print(f'Upper Boundary: {upbound:.2f}' )
print(f'Lower Boundary: {lowbound:.2f}')

# Menghitung persentase outlier
outliers = clean[(clean['gross income'] < lowbound) | (clean['gross income'] > upbound)]
percentage_of_outliers = len(outliers) * 100 / len(clean['gross income'])
print(f'Percentage of outliers: {percentage_of_outliers:.2f}%')
```

dapat disimpulkan bahwa Percantage of Outlier untuk Gross income adalah 0,90%

```python
## Untuk mengetahui nilai outlier pada variabel Total

import pandas as pd

# Membaca file data
data = pd.read_csv('supermarket_sales.csv')

# Membersihkan data dari missing values
clean = data.dropna()

# Mengubah tipe data kolom 'Total' menjadi numerik
clean['Total'] = pd.to_numeric(clean['Total'], errors='coerce')

# Menghitung quartile 1, quartile 3, dan IQR
q1 = clean['Total'].quantile(0.25)
q3 = clean['Total'].quantile(0.75)
iqr = q3 - q1

# Menghitung batas atas dan batas bawah
upbound = q3 + 1.5 * iqr
lowbound = q1 - 1.5 * iqr

print(f'Upper Boundary: {upbound:.2f}' )
print(f'Lower Boundary: {lowbound:.2f}')

# Menghitung persentase outlier
outliers = clean[(clean['Total'] < lowbound) | (clean['Total'] > upbound)]
percentage_of_outliers = len(outliers) * 100 / len(clean['Total'])
print(f'Percentage of outliers: {percentage_of_outliers:.2f}%')
```

dapat disimpulkan bahwa Percantage of Outlier untuk Total adalah 0,90%

## Analisa Methode Statistik Descrptive (Central Tendensi dan Uji Korelasi))

Dimana dalam proses ini akan dihitung rata-rata Pendapatan Dari Setiap Kota yang ditinjau dari beberapa variabel yaitu variabel "Gross income dan Total", kemudian setelah dilakukan anliasa dengan menggunakan central tendensi proses berikutnya yaitu menguji variabel gross income dan total dengan uji korelasi. Berikut adalah studi kasus soal yang perlu di uji dengan uji central tendensi dan uji korelasi

g. Bagaimana Analisa Central tendency data untuk Pendapatan pada setiap kota yang kemudian dilakukan uji korelasi dari bebrapa variabel yang telah diketahui sifat karaktersitik data (memiliki sifat skeww atau tidak? (Metodhe Statistik Descriptive)

```python
## Untuk melihat hasil Central Tendensi total pendapatan ke semua kota

import pandas as pd

# Membaca file data
data = pd.read_csv('supermarket_sales.csv')

# Menghapus baris dengan nilai kosong pada kolom 'gross income'
clean = data.dropna(subset=['gross income'])

# Menampilkan deskripsi statistik dari kolom 'gross income'
gross_income_description = clean['gross income'].describe()

# Menampilkan hasil
print(gross_income_description)
```

Dari hasil diatas dapat disimpulkan bahwa telah didapatkan untuk methode statistika deskriptif berupa nilai rat-rata, standar deviasi, nilai minimum, nilai quartilq serta nilai maksimum dari data Pendapatan (Gross income)dari semua cabang Supemarket Saqinah yang berada di Mandalay, Naypyitaw, Yangon

```python
## Untuk melihat hasil central tendensi dari nilai Pendapatan dari semua kota

import pandas as pd

# Membaca file data
data = pd.read_csv('supermarket_sales.csv')

# Menghapus baris dengan nilai kosong pada kolom 'Pendapatan'
clean = data.dropna(subset=['gross income'])

# Menghitung rata-rata pendapatan per kota
mean_pendapatan = clean.groupby('City')['gross income'].mean()

# Menghitung median pendapatan per kota
median_pendapatan = clean.groupby('City')['gross income'].median()

# Menghitung standar deviasi pendapatan per kota
std_dev_pendapatan = clean.groupby('City')['gross income'].std()

# Menampilkan hasil
for kota in mean_pendapatan.index:
    print(f'City: {kota}')
    print(f'Rata-rata gross income: {mean_pendapatan[kota]:.2f}')
    print(f'Median gross income: {median_pendapatan[kota]:.2f}')
    print(f'Standar Deviasi gross income: {std_dev_pendapatan[kota]:.2f}')
    print()
```

Dari hasil diatas dapat disimpulkan bahwa kota "Naypytiaw" memilki nilai gross income (Pendapatan) paling tinggi dan untuk kota Yangon memilki nilai gross income paling rendah

```python

##Melakukan Uji korelasi antara variabel gross income dengan variabel Total

import pandas as pd
from scipy import stats

# Membaca file data
data = pd.read_csv('supermarket_sales.csv')

# Menghapus baris dengan nilai kosong pada kolom 'gross income' dan 'Total'
clean = data.dropna(subset=['gross income', 'Total'])

# Menghitung korelasi Pearson
corr_r, pval_p = stats.pearsonr(clean['gross income'], clean['Total'])

# Menghitung korelasi Spearman
corr_rho, pval_s = stats.spearmanr(clean['gross income'], clean['Total'])

# Menghitung korelasi Kendall
corr_tau, pval_k = stats.kendalltau(clean['gross income'], clean['Total'])

# Menampilkan hasil
print(f"r-correlation: {corr_r:.2f}, p-value: {pval_p}")
print(f"rho-correlation: {corr_rho:.2f}, p-value: {pval_s}")
print(f"tau-correlation: {corr_tau:.2f}, p-value: {pval_k}")
```

Dalam uji korelasi ini maka dapat disimpulkan bahwa, karena nilai r-correlation adalah 1.00 dan p-value adalah 0,0, ini menunjukkan bahwa adanya hubungan linier positif yang sempurna antara 'gross income' dan 'Total'. Artinya, ketika nilai 'gross income' naik, nilai 'Total' juga naik secara proporsional, begitupun sebaliknya

h.Apakah terdapat perbedaan pembelian product dari konsumen berdasarkan gender ?

Pada uji hipotesis ini, tujuannya adalah untuk menentukan apakah terdapat perbedaan pembelian dari konsumen berdasarkan jenis kelamin. Data yang akan digunakan adalah data gross income yang diperoleh dari perusahaan. Dan adapaun untuk analisa uji hipotesis yang akan dilakukan menggunakan metode Two-Sample Independent t-test (dikarenakan varibael yang akan diuji memiliki lebih dari satu variabe), dengan hipotesis sebagai berikut:

- H0: Tidak ada perbedaan yang signifikan antara rata-rata pembelian produk pelanggan perempuan dan rata-rata pembelian produk pelanggan laki-laki di Supermarket Saqinah.
- H1: Terdapat perbedaan yang signifikan antara rata-rata pembelian produk pelanggan perempuan dan rata-rata pembelian produk pelanggan laki-laki di Supermarket Saqinah.

Setelah dilakukan uji hipotesa maka langkah selanjutnya yaitu melakukan analisa uji chi-kudrat agar dapat diketahui apakah hipotesa ini diterima atau ditolak. Alasan mengapa menggunakan uji chi-kuadrat dikarenakan salah satu variabel data bersifat distribusi tidak normal yaitu variabel Quantity

```python

## Melakukan perhitungan untuk Uji Chi-Square
import pandas as pd
from scipy import stats

#Membaca data dari file CSV
data = pd.read_csv('supermarket_sales.csv')

#Mengambil sampel pembelian produk konsumen perempuan dan laki-laki
perempuan = data[data['Gender'] == 'Female']['Quantity']
lakilaki = data[data['Gender'] == 'Male']['Quantity']

#Melakukan uji Two-Sample Independent t-test
statistic, p_val = stats.ttest_ind(perempuan, lakilaki, equal_var=False)
print('P-value:', p_val)

#Menentukan tingkat signifikansi
alpha = 0.05

#Mengevaluasi hipotesis
if p_val < alpha:
    print('Tolak H0')
else:
    print('Terima H0')
```

Berdasarkan hasil uji chi-square dengan nilai p-value sebesar 0.0189, dengan tingkat signifikansi alpha 0.05, dapat disimpulkan bahwa terdapat perbedaan yang signifikan dalam pembelian produk dari konsumen berdasarkan gender di supermarket. Oleh karena itu, hipotesis nol (H0) yang menyatakan tidak ada perbedaan yang signifikan antara pembelian produk konsumen perempuan dan laki-laki "ditolak".

```python

#Membuat visualisasi diagram violin pembelian barang
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Membaca data dari file CSVsupermarket_sales.csv'
data = pd.read_csv('supermarket_sales.csv')

# Membuat violin plot untuk memvisualisasikan pembelian barang di supermarket Saqinah berdasarkan jenis kelamin
plt.figure(figsize=(10, 6))
sns.set_theme(style="whitegrid")
sns.violinplot(x='Gender', y='Quantity', data=data, hue='Gender', palette='Set2')
plt.title('Violin Plot Pembelian Barang di Supermarket Saqinah Berdasarkan Gender')
plt.xlabel('Gender')
plt.ylabel('Quantity')
plt.legend()
plt.show()
```

i. Apakah ada perbedaan penjualan untuk masing-masing cabang supermarket Saqinah (Mandalay, Naypytay,dan Yangon (Methode Inferensial )

Pada uji hipotesis ini, tujuannya adalah untuk menentukan apakah terdapat perbedaan pembelian dari konsumen berdasarkan jenis kelamin. Data yang akan digunakan adalah data gross income yang diperoleh dari perusahaan. Dan adapaun untuk analisa uji hipotesis yang akan dilakukan menggunakan metode Two-Sample Independent t-test (dikarenakan varibael yang akan diuji memiliki lebih dari satu variabe), dengan hipotesis sebagai berikut:

- Hipotesis Nol (H0): Tidak ada perbedaan signifikan dalam penjualan antara masing-masing cabang supermarket Saqinah.
- Hipotesis Alternatif (H1): Terdapat perbedaan signifikan dalam penjualan antara masing-masing cabang supermarket Saqinah.

Setelah dilakukan uji hipotesa maka langkah selanjutnya yaitu melakukan analisa uji chi-kudrat agar dapat diketahui apakah hipotesa ini diterima atau ditolak. Alasan mengapa menggunakan uji chi-kuadrat dikarenakan salah satu variabel data bersifat distribusi tidak normal yaitu variabel Quantity




```python
# Melakukan pengujian dengan Uji Chi-Kuadrat 

import pandas as pd
from scipy import stats

# Membaca data dari file CSV
data = pd.read_csv('supermarket_sales.csv')

# Mengambil sampel pembelian produk konsumen perempuan dan laki-laki
Mandalay = data[data['City'] == 'Mandalay']['Quantity']
Naypyitaw = data[data['City'] == 'Naypyitaw']['Quantity']
Yangon = data[data['City'] == 'Yangon']['Quantity']

# Melakukan uji Two-Sample Independent t-test
statistic, p_val = stats.ttest_ind(Mandalay, Naypyitaw, equal_var=False)
print('P-value:', p_val)

# Menentukan tingkat signifikansi
alpha = 0.05

# Mengevaluasi hipotesis
if p_val < alpha:
    print('Tolak H0')
else:
    print('Terima H0')
```

Berdasarkan nilai p-value sebesar 0.6632532437958025 dan dengan tingkat signifikansi  = 0.05, kita dapat menyimpulkan bahwa tidak ada cukup bukti statistik yang mendukung perbedaan yang signifikan dalam pembelian produk antara cabang supermarket Saqinah Mandalay, Naypyitaw, dan Yangon. Oleh karena itu, hipotesis nol (H0) "diterima", yang berarti tidak ada perbedaan yang signifikan dalam penjualan antara cabang-cabang tersebut.
